```python
import pandas as pd
import altair as alt
```

```python
df = pd.read_csv('data/age-dependency-ratio-projected-to-2100.csv')
df.head()
```

```python
df = df.rename(columns={'Estimates, 1950 - 2020: Dependency ratios (both sexes combined) for different age groups - Annual total dep. ratio [(0-14 & 65+) / 15-64] (%)': 'ratios'})
world = df[df['Entity'] == 'World']
world.head()
```

```python
alt.Chart(world).mark_line().encode(
    x='Year',
    y='ratios:Q'
).properties(
    height=600,
    width=800
)
```
