```python
nato = df.loc[df['Entity'].isin(['NATO members'])]
## nato is a DF
nato.drop(columns=['Code'])
```

```python
import seaborn as sns
sns.set(rc = {'figure.figsize':(15,8)})
sns.set(style='dark',)
sns.lineplot(x='Year', y='military_expenditure',data=nato)
```

```python
relevant_entity_codes = ['NATO members', 'United States', 'China', 'India', 'Russia', 'France', 'Germany', 'Japan' ,'Ukraine']
relevant_entity_series = df.loc[df['Entity'].isin(relevant_entity_codes)]
relevant_entity_series.drop(columns=['Code'])
```

```python
sns.lineplot(x='Year', y='military_expenditure', hue='Entity', data=relevant_entity_series)
```

```python
def get_world_expenditures_for_year(year):
    world_expenditures_for_year = df.loc[df['Year'] == year]
    return world_expenditures_for_year.drop(columns=['Code'])

we = get_world_expenditures_for_year(1979)
we.head()
```
